```python
total = []

for i in range(150):

    forward = iris[:i]
    reverse = iris[i:]

    forward_medians = np.median(forward,axis=0)
    reverse_medians = np.median(reverse,axis=0)
    
    fsr = np.sum(np.power(forward-forward_medians,2),axis=0)
    rsr = np.sum(np.power(reverse-reverse_medians,2),axis=0)
    
    print("*****************")
    print(i)
    print(fsr)
    print(rsr)
    print(fsr+rsr)
    total.append(np.sum(fsr+rsr))
```
